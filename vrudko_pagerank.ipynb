```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
"""
Authors:
    - Rudko, Viktor
"""
```

```python
plt.figure(1)
DirectedGraph1 = nx.gnm_random_graph(n=4, m=8, seed=523, directed=True)
print(f"Degrees of PR graph 1 with n = {len(DirectedGraph1.nodes)} and m = {len(DirectedGraph1.edges)}:\n")
nx.draw(DirectedGraph1, pos=nx.circular_layout(DirectedGraph1), with_labels=True)
print("Degrees of PR graph 2:\n")
for v in nx.nodes(DirectedGraph1):
    print(f"In-Degree of vertex {v}: {DirectedGraph1.in_degree(v)}")
plt.savefig("Directed_Graph_PR_1.png", format="PNG", dpi=200)
plt.show()

ranks_1 = nx.pagerank(DirectedGraph1)
ranks_1_sorted = sorted(ranks_1.items(), key=lambda x: x[1], reverse=True)
print(f"Pageranks for directed graph 2 with n = {len(DirectedGraph1.nodes)}, m = {len(DirectedGraph1.edges)}:\n{ranks_1_sorted}")
```

```python
plt.figure(3)
DirectedGraph3 = nx.gnm_random_graph(n=5, m=7, seed=543, directed=True)
print(DirectedGraph3)
nx.draw(DirectedGraph3, with_labels=True)
plt.savefig("Directed_Graph_3.png", format="PNG", dpi=200)
plt.show()
DirectedGraph3_additional = DirectedGraph3
for i in range(len(DirectedGraph3_additional.nodes), 15):
    DirectedGraph3_additional.add_node(i)
    DirectedGraph3_additional.add_edge(i, 3)
nx.draw(DirectedGraph3_additional, with_labels=True)
plt.savefig("Directed_Graph_3_additional.png", format="PNG", dpi=200)
plt.show()
ranks_3_add = nx.pagerank(DirectedGraph3_additional)
ranks_3_add_sorted = sorted(ranks_3_add.items(), key=lambda x: x[1], reverse=True)
print(f"Pageranks for directed graph 3 with additional nodes:\n{ranks_3_add_sorted}")
DirectedGraph3_additional_2 = DirectedGraph3_additional
DirectedGraph3_additional_2.remove_edge(2, 4)
DirectedGraph3_additional_2.remove_edge(0, 4)
DirectedGraph3_additional_2.add_edge(2, 3)
nx.draw(DirectedGraph3_additional_2, with_labels=True)
plt.savefig("Directed_Graph_3_additional_modified.png", format="PNG", dpi=200)
plt.show()
ranks_3_add_2 = nx.pagerank(DirectedGraph3_additional_2)
ranks_3_add_2_sorted = sorted(ranks_3_add_2.items(), key=lambda x: x[1], reverse=True)
print(f"Pageranks for directed graph 3 with additional nodes and modifications:\n{ranks_3_add_2_sorted}")

```
