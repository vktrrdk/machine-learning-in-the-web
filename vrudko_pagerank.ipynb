```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
"""
Authors:
    - Rudko, Viktor
"""
```

```python
plt.figure(1)
DirectedGraph1 = nx.gnm_random_graph(n=4, m=8, seed=523, directed=True)
print(f"Degrees of PR graph 1 with n = {len(DirectedGraph1.nodes)} and m = {len(DirectedGraph1.edges)}:\n")
nx.draw(DirectedGraph1, pos=nx.circular_layout(DirectedGraph1), with_labels=True)
print("Degrees of PR graph 2:\n")
for v in nx.nodes(DirectedGraph1):
    print(f"In-Degree of vertex {v}: {DirectedGraph1.in_degree(v)}")
plt.savefig("Directed_Graph_PR_1.svg", format="SVG", dpi=200, transparent=True)
plt.show()

ranks_1 = nx.pagerank(DirectedGraph1)
ranks_1_sorted = sorted(ranks_1.items(), key=lambda x: x[1], reverse=True)
print(f"Pageranks for directed graph 2 with n = {len(DirectedGraph1.nodes)}, m = {len(DirectedGraph1.edges)}:\n{ranks_1_sorted}")
```

```python
plt.figure(3)
DirectedGraph3 = nx.gnm_random_graph(n=5, m=7, seed=543, directed=True)
nx.draw(DirectedGraph3, with_labels=True)
plt.savefig("Directed_Graph_3.svg", format="svg", dpi=200, transparent=True)
plt.show()
ranks_3 = nx.pagerank(DirectedGraph3)
ranks_3_sorted = dict(sorted(ranks_3.items(), key=lambda x: x[1], reverse=True))
for key in ranks_3_sorted:
    ranks_3_sorted[key] = round(ranks_3_sorted[key], 4)
print(f"Pageranks for modified graph:\n{ranks_3_sorted}")
DirectedGraph3_additional = DirectedGraph3
for i in range(len(DirectedGraph3_additional.nodes), 15):
    DirectedGraph3_additional.add_node(i)
    DirectedGraph3_additional.add_edge(i, 3)
DirectedGraph3_additional.remove_edge(2, 4)
DirectedGraph3_additional.remove_edge(0, 4)
DirectedGraph3_additional.remove_edge(3, 4)
DirectedGraph3_additional.add_edge(2, 3)
nx.draw(DirectedGraph3_additional, with_labels=True, pos=nx.planar_layout(DirectedGraph3_additional))
plt.savefig("Directed_Graph_additional.svg", format="svg", dpi=200, transparent=True)
plt.show()
ranks_3_add = nx.pagerank(DirectedGraph3_additional)
ranks_3_add_sorted = dict(sorted(ranks_3_add.items(), key=lambda x: x[1], reverse=True))
for key in ranks_3_add_sorted:
    ranks_3_add_sorted[key] = round(ranks_3_add_sorted[key], 4)
print(f"Pageranks for modified graph:\n{ranks_3_add_sorted}")
DirectedGraph3_modified = DirectedGraph3_additional
DirectedGraph3_modified.add_node(len(DirectedGraph3_modified.nodes))
DirectedGraph3_modified.add_edge(3, len(DirectedGraph3_modified.nodes) -1)
nx.draw(DirectedGraph3_modified, with_labels=True, pos=nx.planar_layout(DirectedGraph3_modified))
plt.savefig("Directed_Graph_modified.svg", format="svg", dpi=200, transparent=True)
plt.show()
ranks_3_mod = nx.pagerank(DirectedGraph3_modified)
ranks_3_mod_sorted = dict(sorted(ranks_3_mod.items(), key=lambda x: x[1], reverse=True))
for key in ranks_3_mod_sorted:
    ranks_3_mod_sorted[key] = round(ranks_3_mod_sorted[key], 4)
print(f"Pageranks for graph with additional node:\n{ranks_3_mod_sorted}")
```
